```python
#22
#1.–°—Ç–≤–æ—Ä—ñ—Ç—å –æ–¥–Ω–æ–≤–∏–º—ñ—Ä–Ω–∏–π –º–∞—Å–∏–≤ (–≤–µ–∫—Ç–æ—Ä) –∑—ñ –∑–Ω–∞—á–µ–Ω–Ω—è–º–∏ ùëíùë•–¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞ x, –ø–æ—á–∏–Ω–∞—é—á–∏ –∑ 1, —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å –π–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è.

import numpy as np

# –í–∏–∑–Ω–∞—á–∞—î–º–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞, –ø–æ—á–∏–Ω–∞—é—á–∏ –∑ 1 (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –¥–æ 10)
x_values = np.arange(1, 11)

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Å–∏–≤ –∑—ñ –∑–Ω–∞—á–µ–Ω–Ω—è–º–∏ e^x
exp_values = np.exp(x_values)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(exp_values)

```

```python
#2.–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤–æ–≤–∏–º—ñ—Ä–Ω–∏–π –º–∞—Å–∏–≤ (–º–∞—Ç—Ä–∏—Ü—é) —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å –π–æ–≥–æ —Ç–∞–Ω–≥–µ–Ω—Å–∞–º–∏ –∫–æ–∂–Ω–æ–≥–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞, –ø–æ—á–∏–Ω–∞—é—á–∏ –∑ 1, —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å –π–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è.

import numpy as np

# –í–∏–∑–Ω–∞—á–∞—î–º–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞ –≤—ñ–¥ 1 –¥–æ 9 (–≤—Å—å–æ–≥–æ 9 –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥–ª—è –º–∞—Ç—Ä–∏—Ü—ñ 3x3)
x_values = np.arange(1, 10)

# –û–±—á–∏—Å–ª—é—î–º–æ —Ç–∞–Ω–≥–µ–Ω—Å –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è
tan_values = np.tan(x_values)

# –§–æ—Ä–º—É—î–º–æ –º–∞—Ç—Ä–∏—Ü—é 3x3
matrix = tan_values.reshape(3, 3)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(matrix)
```

```python
#3.–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å –π–æ–≥–æ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -1 –¥–æ 1 —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞.

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Å–∏–≤ 3x3 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -1 –¥–æ 1
matrix = np.random.uniform(-1, 1, (3, 3))

# –í–∏–≤–æ–¥–∏–º–æ —Å—Ç–≤–æ—Ä–µ–Ω–∏–π –º–∞—Å–∏–≤
print("–°—Ç–≤–æ—Ä–µ–Ω–∏–π –º–∞—Å–∏–≤:\n", matrix)

# –Ø–∫—â–æ —Ç—Ä–µ–±–∞ –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ —Å–µ—Ä–µ–¥ –¥—ñ–π—Å–Ω–∏—Ö (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –∑–∞ –º–æ–¥—É–ª–µ–º)
# –ü–µ—Ä–µ—Ç–≤–æ—Ä–∏–º–æ –¥—ñ–π—Å–Ω—ñ —á–∏—Å–ª–∞ —É —Ü—ñ–ª—ñ —Ç–∞ –ø–µ—Ä–µ–≤—ñ—Ä–∏–º–æ —ó—Ö –ø–∞—Ä–Ω—ñ—Å—Ç—å
odd_numbers = matrix[np.abs(matrix % 2) == 1]

# –í–∏–≤–æ–¥–∏–º–æ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞
print("–ù–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞:\n", odd_numbers)
```

```python
#4.–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä–æ–º 4x4, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å –π–æ–≥–æ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -0.8 –¥–æ 0.8 —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å –π–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Å–∏–≤ 4x4 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -0.8 –¥–æ 0.8
matrix = np.random.uniform(-0.8, 0.8, (4, 4))

# –í–∏–≤–æ–¥–∏–º–æ —Å—Ç–≤–æ—Ä–µ–Ω–∏–π –º–∞—Å–∏–≤
print("–°—Ç–≤–æ—Ä–µ–Ω–∏–π –º–∞—Å–∏–≤:\n", matrix)
```

```python
#5.–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤–∞ –æ–¥–Ω–æ–≤–∏–º—ñ—Ä–Ω–∏—Ö –º–∞—Å–∏–≤–∏ —Ä–æ–∑–º—ñ—Ä–æ–º 6, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—Ö –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 5 –¥–æ 15 —Ç–∞ –≤–∏–∫–æ–Ω–∞–π—Ç–µ –Ω–∞ –Ω–∏—Ö –ø–æ–µ–ª–µ–º–µ–Ω—Ç–Ω—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó –¥–æ–¥–∞–≤–∞–Ω–Ω—è, –≤—ñ–¥–Ω—ñ–º–∞–Ω–Ω—è —Ç–∞ –º–Ω–æ–∂–µ–Ω–Ω—è.

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –¥–≤–∞ –æ–¥–Ω–æ–≤–∏–º—ñ—Ä–Ω–∏—Ö –º–∞—Å–∏–≤–∏ —Ä–æ–∑–º—ñ—Ä–æ–º 6, –∑–∞–ø–æ–≤–Ω–µ–Ω—ñ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 5 –¥–æ 15
array1 = np.random.randint(5, 16, 6)
array2 = np.random.randint(5, 16, 6)

# –ü–æ–µ–ª–µ–º–µ–Ω—Ç–Ω–µ –¥–æ–¥–∞–≤–∞–Ω–Ω—è
addition = array1 + array2

# –ü–æ–µ–ª–µ–º–µ–Ω—Ç–Ω–µ –≤—ñ–¥–Ω—ñ–º–∞–Ω–Ω—è
subtraction = array1 - array2

# –ü–æ–µ–ª–µ–º–µ–Ω—Ç–Ω–µ –º–Ω–æ–∂–µ–Ω–Ω—è
multiplication = array1 * array2

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–µ—Ä—à–∏–π –º–∞—Å–∏–≤:", array1)
print("–î—Ä—É–≥–∏–π –º–∞—Å–∏–≤:", array2)
print("\n–†–µ–∑—É–ª—å—Ç–∞—Ç –¥–æ–¥–∞–≤–∞–Ω–Ω—è:", addition)
print("–†–µ–∑—É–ª—å—Ç–∞—Ç –≤—ñ–¥–Ω—ñ–º–∞–Ω–Ω—è:", subtraction)
print("–†–µ–∑—É–ª—å—Ç–∞—Ç –º–Ω–æ–∂–µ–Ω–Ω—è:", multiplication)
```

```python
#6.–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ 4x4, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—Ö –¥–æ–≤—ñ–ª—å–Ω–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -6 –¥–æ 6 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å —ó—Ö –¥–æ–±—É—Ç–æ–∫

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ 4x4, –∑–∞–ø–æ–≤–Ω–µ–Ω—ñ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -6 –¥–æ 6
matrix1 = np.random.randint(-6, 7, (4, 4))
matrix2 = np.random.randint(-6, 7, (4, 4))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ –¥–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—å
product = np.dot(matrix1, matrix2)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–µ—Ä—à–∞ –º–∞—Ç—Ä–∏—Ü—è:\n", matrix1)
print("–î—Ä—É–≥–∞ –º–∞—Ç—Ä–∏—Ü—è:\n", matrix2)
print("\n–î–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—å:\n", product)
```

```python
#7.–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º 3x2 —Ç–∞ 2x2, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—Ö –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 8 —Ç–∞ –ø–µ—Ä–µ–º–Ω–æ–∂—Ç–µ —ó—Ö –º—ñ–∂ —Å–æ–±–æ—é.

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x2, –∑–∞–ø–æ–≤–Ω–µ–Ω—É –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 8
matrix1 = np.random.randint(1, 9, (3, 2))

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 2x2, –∑–∞–ø–æ–≤–Ω–µ–Ω—É –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 8
matrix2 = np.random.randint(1, 9, (2, 2))

# –ü–µ—Ä–µ–º–Ω–æ–∂—É—î–º–æ –º–∞—Ç—Ä–∏—Ü—ñ
product = np.dot(matrix1, matrix2)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–µ—Ä—à–∞ –º–∞—Ç—Ä–∏—Ü—è (3x2):\n", matrix1)
print("–î—Ä—É–≥–∞ –º–∞—Ç—Ä–∏—Ü—è (2x2):\n", matrix2)
print("\n–î–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—å (3x2 * 2x2):\n", product)
```

```python
#8.–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—ó –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 8 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å —ó—ó –æ–±–µ—Ä–Ω–µ–Ω—É –º–∞—Ç—Ä–∏—Ü—é.

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω–µ–Ω—É –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 8
matrix = np.random.randint(1, 9, (3, 3))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ –æ–±–µ—Ä–Ω–µ–Ω—É –º–∞—Ç—Ä–∏—Ü—é
try:
    inverse_matrix = np.linalg.inv(matrix)
except np.linalg.LinAlgError:
    inverse_matrix = "–ú–∞—Ç—Ä–∏—Ü—è —î –≤–∏—Ä–æ–¥–∂–µ–Ω–æ—é —ñ –Ω–µ –º–∞—î –æ–±–µ—Ä–Ω–µ–Ω–æ—ó."

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–°—Ç–≤–æ—Ä–µ–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):\n", matrix)
print("\n–û–±–µ—Ä–Ω–µ–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è:\n", inverse_matrix)
```

```python
#9.–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 4x4, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—ó –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ -1 –¥–æ 1 —Ç–∞ —Ç—Ä–∞–Ω—Å–ø–æ–Ω—É–π—Ç–µ —ó—ó

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 4x4, –∑–∞–ø–æ–≤–Ω–µ–Ω—É –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -1 –¥–æ 1
matrix = np.random.uniform(-1, 1, (4, 4))

# –¢—Ä–∞–Ω—Å–ø–æ–Ω—É—î–º–æ –º–∞—Ç—Ä–∏—Ü—é
transposed_matrix = np.transpose(matrix)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–°—Ç–≤–æ—Ä–µ–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è (4x4):\n", matrix)
print("\n–¢—Ä–∞–Ω—Å–ø–æ–Ω–æ–≤–∞–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è:\n", transposed_matrix)
```

```python
#10.–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x4 —Ç–∞ –≤–µ–∫—Ç–æ—Ä —Ä–æ–∑–º—ñ—Ä–æ–º 4x1, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é–≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 0 –¥–æ 1, –∞ –≤–µ–∫—Ç–æ—Ä - –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 10, —ñ –∑–Ω–∞–π–¥—ñ—Ç—å –¥–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä.

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x4, –∑–∞–ø–æ–≤–Ω–µ–Ω—É –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 0 –¥–æ 1
matrix = np.random.uniform(0, 1, (3, 4))

# –°—Ç–≤–æ—Ä—é—î–º–æ –≤–µ–∫—Ç–æ—Ä —Ä–æ–∑–º—ñ—Ä–æ–º 4x1, –∑–∞–ø–æ–≤–Ω–µ–Ω–∏–π –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 10
vector = np.random.randint(1, 11, (4, 1))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ –¥–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä
product = np.dot(matrix, vector)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–°—Ç–≤–æ—Ä–µ–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è (3x4):\n", matrix)
print("\n–°—Ç–≤–æ—Ä–µ–Ω–∏–π –≤–µ–∫—Ç–æ—Ä (4x1):\n", vector)
print("\n–î–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä:\n", product)
```

```python
#11.–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 2x3 —Ç–∞ –≤–µ–∫—Ç–æ—Ä —Ä–æ–∑–º—ñ—Ä–æ–º 3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 10, –∞ –≤–µ–∫—Ç–æ—Ä - –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 0 –¥–æ 1, —ñ –∑–Ω–∞–π–¥—ñ—Ç—å –¥–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 2x3, –∑–∞–ø–æ–≤–Ω–µ–Ω—É –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 10
matrix = np.random.randint(1, 11, (2, 3))

# –°—Ç–≤–æ—Ä—é—î–º–æ –≤–µ–∫—Ç–æ—Ä —Ä–æ–∑–º—ñ—Ä–æ–º 3, –∑–∞–ø–æ–≤–Ω–µ–Ω–∏–π –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 0 –¥–æ 1
vector = np.random.uniform(0, 1, 3)

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ –¥–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä
product = np.dot(matrix, vector)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–°—Ç–≤–æ—Ä–µ–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è (2x3):\n", matrix)
print("\n–°—Ç–≤–æ—Ä–µ–Ω–∏–π –≤–µ–∫—Ç–æ—Ä (3):\n", vector)
print("\n–î–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä:\n", product)
```

```python
#12.–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—Ö –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 3 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å —Å—É–º—É –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥—Ä—É–≥–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ.
import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω–µ–Ω—ñ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 3
matrix1 = np.random.randint(1, 4, (3, 3))
matrix2 = np.random.randint(1, 4, (3, 3))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ —Å—É–º—É –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥—Ä—É–≥–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ
sum_matrix2 = np.sum(matrix2)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–µ—Ä—à–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):\n", matrix1)
print("\n–î—Ä—É–≥–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):\n", matrix2)
print("\n–°—É–º–∞ –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥—Ä—É–≥–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ:", sum_matrix2)
```

```python
#13.–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—Ö –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 5 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å —Å—É–º—É —ó—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤.

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω–µ–Ω—ñ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 5
matrix1 = np.random.randint(1, 6, (3, 3))
matrix2 = np.random.randint(1, 6, (3, 3))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ —Å—É–º—É –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –æ–±–æ—Ö –º–∞—Ç—Ä–∏—Ü—å
sum_matrix1 = np.sum(matrix1)
sum_matrix2 = np.sum(matrix2)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–µ—Ä—à–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):\n", matrix1)
print("\n–î—Ä—É–≥–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):\n", matrix2)
print("\n–°—É–º–∞ –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –ø–µ—Ä—à–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ:", sum_matrix1)
print("–°—É–º–∞ –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥—Ä—É–≥–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ:", sum_matrix2)
```

```python
#14.–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—ó –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 10 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å —Å—É–º—É –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –Ω–∞ –≥–æ–ª–æ–≤–Ω—ñ–π –¥—ñ–∞–≥–æ–Ω–∞–ª—ñ.

import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω–µ–Ω—É –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 10
matrix = np.random.randint(1, 11, (3, 3))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ —Å—É–º—É –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –Ω–∞ –≥–æ–ª–æ–≤–Ω—ñ–π –¥—ñ–∞–≥–æ–Ω–∞–ª—ñ
diagonal_sum = np.trace(matrix)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–°—Ç–≤–æ—Ä–µ–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):\n", matrix)
print("\n–°—É–º–∞ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –Ω–∞ –≥–æ–ª–æ–≤–Ω—ñ–π –¥—ñ–∞–≥–æ–Ω–∞–ª—ñ:", diagonal_sum)
```

```python
#15.–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—ó –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ -1 –¥–æ 1 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å –≤–µ–∫—Ç–æ—Ä-—Å—Ç–æ–≤–ø—á–∏–∫, —â–æ –º—ñ—Å—Ç–∏—Ç—å –¥–æ–±—É—Ç–æ–∫ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –∫–æ–∂–Ω–æ–≥–æ —Å—Ç–æ–≤–ø—Ü—è –º–∞—Ç—Ä–∏—Ü—ñ.
import numpy as np

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω–µ–Ω—É –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -1 –¥–æ 1
matrix = np.random.uniform(-1, 1, (3, 3))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ –¥–æ–±—É—Ç–æ–∫ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –∫–æ–∂–Ω–æ–≥–æ —Å—Ç–æ–≤–ø—Ü—è
column_products = np.prod(matrix, axis=0)

# –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ –≤–µ–∫—Ç–æ—Ä-—Ä—è–¥–æ–∫ –≤ –≤–µ–∫—Ç–æ—Ä-—Å—Ç–æ–≤–ø—á–∏–∫
column_products = column_products.reshape(-1, 1)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–°—Ç–≤–æ—Ä–µ–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):\n", matrix)
print("\n–í–µ–∫—Ç–æ—Ä-—Å—Ç–æ–≤–ø—á–∏–∫, —â–æ –º—ñ—Å—Ç–∏—Ç—å –¥–æ–±—É—Ç–æ–∫ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –∫–æ–∂–Ω–æ–≥–æ —Å—Ç–æ–≤–ø—Ü—è:\n", column_products)
```
